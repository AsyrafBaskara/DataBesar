```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/studentAssessment.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Menampilkan 5 row awal pada .csv
df.head()
```

```python
# Mengecek value yang kosong pada .csv
df.isnull().sum()
```

```python
# Set style
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(20,20))
```

```python
# 1. Distribusi Skor Asesmen
sns.histplot(df['score'], bins=30, kde=True)
plt.title('Distribusi Skor Asesmen')
plt.xlabel('Skor')
plt.ylabel('Frekuensi')
plt.show()

# Menunjukkan sebaran skor yang diperoleh mahasiswa dalam semua asesmen.
```

```python
# 2. Jumlah Pengumpulan Asesmen berdasarkan Tanggal
sns.countplot(data=df, x='date_submitted')
plt.title('Jumlah Pengumpulan Asesmen berdasarkan Tanggal')
plt.xlabel('Tanggal Pengumpulan')
plt.ylabel('Jumlah Pengumpulan')
plt.show()

# Memberikan gambaran mengenai frekuensi pengumpulan asesmen pada setiap tanggal.
```

```python
# 3. Jumlah Pengumpulan untuk Setiap Asesmen
sns.countplot(data=df, x='id_assessment', order=df['id_assessment'].value_counts().index)
plt.title('Jumlah Pengumpulan untuk Setiap Asesmen')
plt.xlabel('ID Asesmen')
plt.ylabel('Jumlah Pengumpulan')
plt.xticks(rotation=90)
plt.show()

# Menunjukkan jumlah pengumpulan untuk setiap asesmen.
```

```python
# 4. Rata-Rata Skor berdasarkan ID Asesmen
sns.barplot(data=df, x='id_assessment', y='score')
plt.title('Rata-Rata Skor berdasarkan ID Asesmen')
plt.xlabel('ID Asesmen')
plt.ylabel('Rata-Rata Skor')
plt.show()

# Menampilkan rata-rata skor yang diperoleh mahasiswa untuk setiap asesmen.
```

```python
# 5. Perbandingan Skor Antara Asesmen yang "Banked" dan Tidak
sns.boxplot(data=df, x='is_banked', y='score')
plt.title('Perbandingan Skor Antara Asesmen yang "Banked" dan Tidak')
plt.xlabel('Is Banked')
plt.ylabel('Skor')
plt.show()

# Menunjukkan distribusi skor antara asesmen yang "banked" dan tidak.
```
