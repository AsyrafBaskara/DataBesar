```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/vle.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Menampilkan 5 row awal pada .csv
df.head()
```

```python
# Mengecek value yang kosong pada .csv
df.isnull().sum()
```

```python
# Set style
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(20,20))
```

```python
# 1. Jumlah Materi VLE per Modul
sns.countplot(data=df, x='code_module')
plt.title('Jumlah Materi VLE per Modul')
plt.xlabel('Kode Modul')
plt.ylabel('Jumlah Materi VLE')
plt.tight_layout()
plt.show()

# Menunjukkan jumlah materi VLE untuk setiap modul.
```

```python
# 2. Distribusi Jenis Aktivitas VLE
sns.countplot(data=df, x='activity_type')
plt.title('Distribusi Jenis Aktivitas VLE')
plt.xlabel('Jenis Aktivitas')
plt.ylabel('Jumlah Aktivitas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menunjukkan distribusi jenis aktivitas VLE.
```

```python
# 3. Jumlah Materi VLE per Minggu
sns.countplot(data=df, x='week_from', order=sorted(df['week_from'].unique()))
plt.title('Jumlah Materi VLE per Minggu')
plt.xlabel('Minggu')
plt.ylabel('Jumlah Materi VLE')
plt.tight_layout()
plt.show()

# Menunjukkan jumlah materi VLE yang tersedia setiap minggu.
```

```python
# 4. Durasi Materi VLE
df['duration'] = df['week_to'] - df['week_from']
sns.histplot(data=df, x='duration', bins=30)
plt.title('Durasi Materi VLE')
plt.xlabel('Durasi (minggu)')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

# Menunjukkan seberapa lama materi VLE biasanya tersedia.
```

```python
# 5. Jumlah Aktivitas per Jenis dan Modul
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='code_module', hue='activity_type')
plt.title('Jumlah Aktivitas per Jenis dan Modul')
plt.xlabel('Kode Modul')
plt.ylabel('Jumlah Aktivitas')
plt.legend(title='Jenis Aktivitas', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

# Menunjukkan jumlah aktivitas berdasarkan jenis dan modul.
```
