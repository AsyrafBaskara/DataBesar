```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
data = pd.read_csv("Dataset/studentVle.csv")
```

```python
# Menampilkan informasi dasar tentang dataset
print(data.info())
```

```python
# Menampilkan lima baris pertama data
print(data.head())
```

```python
# Statistik deskriptif untuk kolom numerik
print(data.describe())
```

```python
# Menghitung jumlah nilai yang hilang
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Visualisasi hubungan antara "sum_click" dan "date"
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="date", y="sum_click")
plt.title("Hubungan antara date dan sum_click")
plt.xlabel("date")
plt.ylabel("sum_click")
plt.show()
```

```python
# Penjelasan :

# Pada diagram diatas kita ditampilkan scatter plot dari hubungan
# antara "sum_click" dan "date"
```

```python
# Visualisasi hubungan antara "sum_click" dan "code_module"
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="code_module", y="sum_click")
plt.title("Hubungan antara code_module dan sum_click")
plt.xlabel("code_module")
plt.ylabel("sum_click")
plt.show()
```
