```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/studentVle.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Menampilkan 5 row awal pada .csv
df.head()
```

```python
# Mengecek value yang kosong pada .csv
df.isnull().sum()
```

```python
# Set style
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(20,20))
```

```python
# 1. Aktivitas Mahasiswa Selama Waktu
sns.lineplot(data=df, x='date', y='sum_click', estimator='sum')
plt.title('Total Klik Mahasiswa Selama Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Total Klik')
plt.tight_layout()
plt.show()

# Menunjukkan total klik mahasiswa sepanjang waktu.
```

```python
# 2. Total Klik per Modul
sns.barplot(data=df, x='code_module', y='sum_click', estimator=sum)
plt.title('Total Klik per Modul')
plt.xlabel('Modul')
plt.ylabel('Total Klik')
plt.tight_layout()
plt.show()

# Menunjukkan total klik per modul.
```

```python
# 3. Rata-Rata Klik per Mahasiswa
avg_click_per_student = df.groupby('id_student')['sum_click'].mean()
sns.histplot(avg_click_per_student, bins=30)
plt.title('Rata-Rata Klik per Mahasiswa')
plt.xlabel('Rata-Rata Klik')
plt.ylabel('Jumlah Mahasiswa')
plt.tight_layout()
plt.show()

# Distribusi rata-rata klik per mahasiswa.
```

```python
# 4. Perbandingan Aktivitas antar Modul
sns.boxplot(data=df, x='code_module', y='sum_click')
plt.title('Perbandingan Aktivitas antar Modul')
plt.xlabel('Modul')
plt.ylabel('Klik')
plt.tight_layout()
plt.show()

# Menunjukkan sebaran klik antar modul.
```

```python
# 5. Aktivitas Mahasiswa per ID Situs
sns.scatterplot(data=df, x='id_site', y='sum_click')
plt.title('Aktivitas Mahasiswa per ID Situs')
plt.xlabel('ID Situs')
plt.ylabel('Klik')
plt.tight_layout()
plt.show()

# Menunjukkan aktivitas mahasiswa berdasarkan ID situs.
```
