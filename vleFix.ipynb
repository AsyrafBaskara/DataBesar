```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca dataset
data = pd.read_csv("Dataset/vle.csv")
```

```python
# Tampilkan 5 baris pertama
print(data.head())

# Informasi tentang dataset
print(data.info())
```

```python
# Statistik deskriptif untuk kolom kategorikal
print(data.describe(include='object'))
```

```python
# Periksa data yang hilang
missing_data = data.isnull().sum()
print(missing_data)
```

```python
# Menghapus baris dengan nilai yang hilang
data = data.dropna()
```

```python
# Mengganti nilai yang hilang dengan rata-rata kolom
data['week_from'].fillna(data['week_from'].mean(), inplace=True)
data['week_to'].fillna(data['week_to'].mean(), inplace=True)
```

```python
# Plot jumlah aktivitas berdasarkan jenis aktivitas
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='activity_type')
plt.title('Jumlah Aktivitas berdasarkan Jenis Aktivitas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Penjelasan :

# Pada diagram diatas kita ditampilkan count plot dari jumlah aktivitas 
# berdasarkan jenis aktivitas yang diolah dari "activity_type" dan "count"
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='code_module')
plt.title('Distribusi Jenis Modul Kode')
plt.xticks(rotation=45)
plt.show()
```

```python
# Penjelasan :

# Pada diagram diatas kita ditampilkan count plot dari distribusi jenis
# modul kode yang didasari oleh "code_module" dan "count"
```
