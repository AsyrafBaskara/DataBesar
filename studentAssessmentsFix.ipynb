```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/studentAssessment.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Count the number of presentations for each module
module_counts = df['code_module'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['score'], kde=True, bins=10)
plt.title('Distribusi Skor')
plt.xlabel('Skor')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Penjelasan :

# Dari Diagram diatas, kita bisa mengetahui bagaimana Distribusi Skor
# yang bisa kita lihat dari xlabel berupa "Skor" dan ylabel berupa "Frekuensi"
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antara Kolom')
plt.show()
```

```python
# Penjelasan :

# Pada heatmap diatas, kita bisa melihat korelasi dari setiap kolom 
# yang ada pada dataset tersebut
```
