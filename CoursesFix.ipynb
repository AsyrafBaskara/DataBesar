```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/courses.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Menampilkan 5 row awal pada .csv
df.head()
```

```python
# Mengecek value yang kosong pada .csv
df.isnull().sum()
```

```python
# Set style
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(20,20))
```

```python
# Subplot 1: Jumlah Kursus berdasarkan Modul
sns.countplot(x='code_module', data=df, palette='viridis')
plt.title('Jumlah Kursus berdasarkan Modul')
plt.xlabel('Kode Modul')
plt.ylabel('Jumlah')
plt.tight_layout()
plt.show()

# Diagram batang yang menunjukkan jumlah kursus untuk setiap modul.
```

```python
# Subplot 2: Distribusi Durasi Presentasi Modul
sns.histplot(df['module_presentation_length'], bins=30, kde=True, color='blue')
plt.title('Distribusi Durasi Presentasi Modul')
plt.xlabel('Durasi Presentasi Modul (hari)')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

# Histogram yang menunjukkan distribusi durasi presentasi modul.
```

```python
# Subplot 3: Jumlah Kursus per Presentasi
sns.countplot(y='code_presentation', data=df, palette='viridis')
plt.title('Jumlah Kursus per Presentasi')
plt.xlabel('Jumlah')
plt.ylabel('Kode Presentasi')
plt.tight_layout()
plt.show()

# Diagram batang yang menunjukkan jumlah kursus untuk setiap presentasi.
```

```python
# Subplot 4: Perbandingan Durasi Presentasi Modul per Kode Presentasi
sns.boxplot(x='code_presentation', y='module_presentation_length', data=df, palette='viridis')
plt.title('Perbandingan Durasi Presentasi Modul per Kode Presentasi')
plt.xlabel('Kode Presentasi')
plt.ylabel('Durasi Presentasi Modul (hari)')
plt.tight_layout()
plt.show()

# Box plot yang menunjukkan distribusi durasi presentasi 
# modul untuk setiap kode presentasi.
```
