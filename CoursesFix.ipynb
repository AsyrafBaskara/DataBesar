```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca .csv
df = pd.read_csv('Dataset/courses.csv')
```

```python
# Membaca Informasi basic mengenai .csv
df.info()
```

```python
# Menampilkan 5 row awal pada .csv
df.head()
```

```python
# Mengecek value yang kosong pada .csv
df.isnull().sum()
```

```python
# Count the number of presentations for each module
module_counts = df['code_module'].value_counts()
```

```python
# Visualisasi dari jumlah presentasi modul per-kode modul
plt.figure(figsize=(4, 4))
module_counts.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Jumlah Presentasi Modul per Kode Modul')
plt.xlabel('Kode Modul')
plt.ylabel('Jumlah Presentasi Modul')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Penjelasan :

# Diagram batang ini menggambarkan frekuensi presentasi untuk setiap modul. 
# Dengan informasi ini, kita dapat memahami modul mana yang paling sering diajarkan
# atau memiliki variasi presentasi terbanyak. Ini bisa menjadi indikator tentang popularitas, 
# kebutuhan, atau pentingnya modul tersebut dalam kurikulum.
```

```python
# Visualisasi dari distribusi durasi presentasi modul berdasarkan kode modul
plt.figure(figsize=(6, 3))
sns.boxplot(x=df['code_module'], y=df['module_presentation_length'], palette="pastel")
plt.title('Distribusi Durasi Presentasi Modul Berdasarkan Kode Modul')
plt.xlabel('Kode Modul')
plt.ylabel('Durasi Presentasi Modul (hari)')
plt.tight_layout()
plt.show()
```

```python
# Penjelasan :

# Visualisasi ini memberikan gambaran tentang variasi durasi presentasi modul antara modul yang berbeda. 
# Dengan membandingkan boxplot untuk setiap modul, kita dapat dengan mudah melihat modul mana yang memiliki durasi 
# presentasi yang lebih panjang atau pendek, serta modul mana yang memiliki variasi durasi yang lebih besar.
```
